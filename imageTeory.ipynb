```python
from skimage.io import imread
```

```python
def get_hist(img):
    hist = [0 for i in range(256)]
    
    for i in img:
        for j in i:
            hist[int(j)] += 1
            
    return hist
```

```python
import matplotlib.pyplot as plt
f, axes = plt.subplots(3, figsize=(10, 10))

img = imread("https://cv1.litres.ru/pub/c/elektronnaya-kniga/cover_415/25440816-mark-menson-tonkoe-iskusstvo-pofigizma.jpg")

axes[0].plot(get_hist(img[:, :, 0]),)
axes[1].plot(get_hist(img[:, :, 1]),)
axes[2].plot(get_hist(img[:, :, 2]),)
```
